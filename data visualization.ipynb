```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
scenarios = ['EoL + low-tech sorting', 'EoL + med-tech sorting', 'EoL + high-tech sorting', 
             'Renewable', 'Electrification', 'Renewable + Electrification', 'All', 
             'Base Case: Linear Economy']
gwp_values = [2.5, 2.06, 1.81, 1.62, 2.67, 1.17, 0.3, 2.68]

# Set up the plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=gwp_values, y=scenarios, palette="coolwarm")

# Adding labels and title
plt.xlabel('GWP (kg CO2 eq. / kg of product)', fontsize=14)
plt.ylabel('Scenario', fontsize=14)

# Adjusting y-axis text size
plt.yticks(fontsize=14)

# Annotating the bars with the values
for index, value in enumerate(gwp_values):
    ax.text(value + 0.05, index, f'{value:.2f}', va='center', fontsize=12)

# Save the plot as a PDF
plt.tight_layout()
plt.savefig('GWP_by_Scenario.pdf', format='pdf')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
scenarios = ['EoL + low-tech sorting', 'EoL + med-tech sorting', 'EoL + high-tech sorting', 
             'Renewable', 'Electrification', 'Renewable + Electrification', 'All', 
             'Base Case: Linear Economy']
gwp_values = [2.5, 2.06, 1.81, 1.62, 2.67, 1.17, 0.3, 2.68]
circularity_values = [0.64, 0.67, 0.79, 0, 0, 0, 0.79, 0]

# Set up the plot
plt.figure(figsize=(10, 6))

# Create scatter plot with different markers for each scenario
for i, scenario in enumerate(scenarios):
    plt.scatter(gwp_values[i], circularity_values[i], label=scenario, s=100)

# Adding labels and title with larger font sizes
plt.xlabel('GWP (kg CO2 eq / kg of product)', fontsize=16)
plt.ylabel('Circularity', fontsize=16)

# Increase the font size of tick marks
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Show legend with larger font
plt.legend(title='Scenario', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12, title_fontsize=14)

# Save the plot as a PDF
plt.tight_layout()
plt.savefig('GWP_Circularity.pdf', format='pdf')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
scenarios = ['EoL + low-tech sorting', 'EoL + med-tech sorting', 'EoL + high-tech sorting', 
             'Renewable', 'Electrification', 'Renewable + Electrification', 'All', 
             'Base Case: Linear Economy']
electricity = [2.14*3.6, 2.19*3.6, 2.87*3.6, 2.17*3.6, 2.67*3.6, 2.67*3.6, 3.11*3.6, 2.17*3.6]
heat = [2.14, 1.39, 0.44, 1.79, 0, 0, 0, 1.79]
total_energy = [9.9, 9.3, 10.8, 9.6, 9.6, 9.6, 11.2, 9.6]

# Set up the bar plot
fig, ax = plt.subplots(figsize=(12, 6))

# Set position for bars
bar_width = 0.25
index = np.arange(len(scenarios))

# Plot bars
bar1 = plt.bar(index, electricity, bar_width, label='Electricity', color='skyblue')
bar2 = plt.bar(index + bar_width, heat, bar_width, label='Heat', color='orange')
bar3 = plt.bar(index + 2 * bar_width, total_energy, bar_width, label='Total Energy', color='green')

# Adding labels and title
plt.ylabel('MJ/kg of product)', fontsize=14)

# Setting the x-axis ticks
plt.xticks(index + bar_width, scenarios, rotation=45, ha='right', fontsize=12)

# Add legend
plt.legend()

# Save the plot as a PDF
plt.tight_layout()
plt.savefig('energy.pdf', format='pdf')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
materials = ['PET', 'HDPE', 'LDPE/LLDPE', 'PP', 'PS', 'PVC']
values = [0.213877362, 0.2031855717, 0.2908748746, 0.2212446645, 0.04990293683, 0.02091459035]

# Convert values to percentages
total = sum(values)
percentages = [(value / total) * 100 for value in values]

# Set up the pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=materials, autopct='%1.1f%%', startangle=90, 
        colors=sns.color_palette('pastel'), textprops={'fontsize': 14})

# Save the plot as a PDF
plt.tight_layout()
plt.savefig('polymer-shares.pdf', format='pdf')

# Show the chart
plt.tight_layout()
plt.show()
```
